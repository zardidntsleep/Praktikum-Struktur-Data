```python
#Representasi Graf dengan Adjacency List

#Membuat graf tak berarah menggunakan adjacency list
graph = {
    'A':['B','C'],
    'B':['A','D','E'],
    'C':['A','F'],
    'D':['B'],
    'E':['B','F'],
    'F':['C','E']
}

#Menampilkan graf
def print_graph(graf):
  for node in graf:
    print(f"{node} --> {graf[node]}")

print_graph(graph)
```

```python
#Traversal Graf menggunakan BFS

from collections import deque

def bfs(graph, start):
  visited = set()
  queue = deque ([start])

  while queue:
    vertex = queue.popleft()
    if vertex not in visited:
      print(vertex, end=" ")
      visited.add(vertex)
      queue.extend([neighbor for neighbor in graph [vertex] if neighbor not in visited])

#Contoh pemanggilan
bfs(graph, 'A')
```

```python
#Traversal Graf menggunakan DFS

def dfs(graph, start, visited=None):
  if visited is None:
    visited = set()

  if start not in visited:
    print(start, end=" ")
    visited.add(start)
    for neighbor in graph[start]:
      dfs(graph, neighbor, visited)

#Contoh pemanggilan
dfs(graph, 'A')
```

```python
#Lengkapi fungsi untuk menghitung jumlah tetangga dari tiap simpul dalam graf
def count_neighbors(graph):
    result = {}

    for node in graph:
        #Hitung jumlah tetangga
        result[node] = len(graph[node])

    return result

#Pemanggilan fungsi
hasil = count_neighbors(graph)

#Menampilkan hasil
print("Jumlah tetangga tiap simpul:")
for node in hasil:
    print(f"{node} = {hasil[node]}")
```

```python
from collections import deque

#Graf utama
graph = {
    'A': ['B', 'C', 'D'],
    'B': ['A', 'C', 'E'],
    'C': ['A', 'B', 'F'],
    'D': ['A', 'E'],
    'E': ['B', 'D', 'F', 'G'],
    'F': ['C', 'E', 'G'],
    'G': ['E', 'F'],
}

#1. Graf tak berarah dengan min 5 simpul dan 7 sisi
def print_graph(graph):
    for node in graph:
        print(f"{node} -> {graph[node]}")

print_graph(graph)
```

```python
#2. Implementasi BFS dan DFS
def bfs(graph, start):
    visited = set()
    queue = deque([start])
    while queue:
        vertex = queue.popleft()
        if vertex not in visited:
            print(vertex, end=" ")
            visited.add(vertex)
            queue.extend([n for n in graph[vertex] if n not in visited])

def dfs(graph, start, visited=None):
    if visited is None:
        visited = set()
    if start not in visited:
        print(start, end=" ")
        visited.add(start)
        for neighbor in graph[start]:
            dfs(graph, neighbor, visited)

print("BFS: "); bfs(graph, 'A')
print("\nDFS: "); dfs(graph, 'A')
```

```python
#3. Modifikasi BFS utk mengembalikan list urutan kunjungan
def bfs_list(graph, start):
    visited = set()
    queue = deque([start])
    result = []
    while queue:
        vertex = queue.popleft()
        if vertex not in visited:
            result.append(vertex)
            visited.add(vertex)
            queue.extend([n for n in graph[vertex] if n not in visited])
    return result

print("BFS list:", bfs_list(graph, 'A'))
```

```python
#4. Fungsi find_path yg mengembalikan 1 jalur menggunakan DFS
def find_path(graph, start, end, visited=None, path=None):
    if visited is None:
        visited = set()
        path = []
    visited.add(start)
    path = path + [start]
    if start == end:
        return path
    for neighbor in graph[start]:
        if neighbor not in visited:
            result = find_path(graph, neighbor, end, visited, path)
            if result:
                return result
    return None

print("find_path(A, G):", find_path(graph, 'A', 'G'))
```

```python
#5. Tambah fungsi is_connected
def is_connected(graph):
    visited = set()
    def dfs_visit(node):
        visited.add(node)
        for neighbor in graph[node]:
            if neighbor not in visited:
                dfs_visit(neighbor)
    dfs_visit(list(graph.keys())[0])
    return len(visited) == len(graph)

print("is_connected:", is_connected(graph))
```

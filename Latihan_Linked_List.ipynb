```python
#01. DOUBLE LINKED LIST

class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.nextval = None
        self.prevval = None


class DoubleLinkedList:
    def __init__(self):
        self.headval = None

#print linked list
    def listprint(self):
        printval = self.headval
        while printval is not None:
            print(printval.dataval)
            printval = printval.nextval

#tambah node di akhir
    def AddInEnd(self, newdata):
        NewNode = Node(newdata)
        if self.headval is None:
            self.headval = NewNode
            return

        last = self.headval
        while last.nextval:
            last = last.nextval

        last.nextval = NewNode
        NewNode.prevval = last

#sisip di tengah (setelah key tertentu)
    def InsertAfter(self, key, newdata):
        temp = self.headval

        while temp is not None:
            if temp.dataval == key:
                break
            temp = temp.nextval

        if temp is None:
            print("Key tidak ditemukan")
            return

        NewNode = Node(newdata)
        NewNode.nextval = temp.nextval
        NewNode.prevval = temp

        if temp.nextval is not None:
            temp.nextval.prevval = NewNode

        temp.nextval = NewNode

#hapus berdasarkan kata kunci
    def RemoveByKey(self, key):
        temp = self.headval

        while temp is not None:
            if temp.dataval == key:
                break
            temp = temp.nextval

        if temp is None:
            print("Data tidak ditemukan")
            return

        # jika node pertama
        if temp.prevval is None:
            self.headval = temp.nextval
            if self.headval is not None:
                self.headval.prevval = None
        else:
            temp.prevval.nextval = temp.nextval
            if temp.nextval is not None:
                temp.nextval.prevval = temp.prevval

        temp = None


Li = DoubleLinkedList()

#3 node awal
Li.AddInEnd("Januari")
Li.AddInEnd("Februari")
Li.AddInEnd("Maret")

print("Sebelum penyisipan:")
Li.listprint()
print("========================================")

#sisip tengah (setelah februari)
Li.InsertAfter("Februari", "April")

print("Setelah sisip April setelah Februari:")
Li.listprint()
print("========================================")

#hapus berdasarkan kata kunci
Li.RemoveByKey("Februari")

print("Setelah hapus Februari:")
Li.listprint()
```

```python
#02. CIRCULAR LINKED LIST

class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.nextval = None


class CircularLinkedList:
    def __init__(self):
        self.headval = None

#print circular linked list
    def listprint(self):
        if self.headval is None:
            return

        printval = self.headval
        while True:
            print(printval.dataval)
            printval = printval.nextval
            if printval == self.headval:
                break

#tambah node di akhir
    def AddInEnd(self, newdata):
        NewNode = Node(newdata)

        if self.headval is None:
            self.headval = NewNode
            NewNode.nextval = self.headval
            return

        last = self.headval
        while last.nextval != self.headval:
            last = last.nextval

        last.nextval = NewNode
        NewNode.nextval = self.headval

#hapus node di akhir
    def RemoveEnd(self):
        if self.headval is None:
            return

        #jika hanya 1 node
        if self.headval.nextval == self.headval:
            self.headval = None
            return

        prev = None
        last = self.headval

        while last.nextval != self.headval:
            prev = last
            last = last.nextval

        prev.nextval = self.headval
        last = None


Li = CircularLinkedList()

#3 node awal
Li.AddInEnd("Januari")
Li.AddInEnd("Februari")
Li.AddInEnd("Maret")

print("Sebelum penyisipan:")
Li.listprint()
print("==============================")

#sisip akhir
Li.AddInEnd("April")

print("Setelah tambah di akhir:")
Li.listprint()
print("==============================")

#hapus akhir
Li.RemoveEnd()

print("Setelah hapus node terakhir:")
Li.listprint()
```

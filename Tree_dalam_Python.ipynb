```python
#MEMBUAT BINARY TREE SECARA MANUAL

class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

#Membuat tree secara manual
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)

#Fungsi inorder traversal
def inorder(node):
    if node:
        inorder(node.left)
        print(node.data, end=' ')
        inorder(node.right)

print("Hasil inorder traversal dari binary tree:")
inorder(root)
```

```python
#MEMBUAT BINARY SEARCH TREE

class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

class BinarySearchTree:
    def __init__(self):
        self.root = None

    def insert(self, root, data):
        if root is None:
            return Node(data)
        if data < root.data:
            root.left = self.insert(root.left, data)
        else:
            root.right = self.insert(root.right, data)
        return root

bst = BinarySearchTree()
root = None
data_list = [50, 30, 70, 20, 40, 60, 80]
for value in data_list:
    root = bst.insert(root, value)
```

```python
#TRAVERSAL PADA BST

def inorder(node):
    if node:
        inorder(node.left)
        print(node.data, end=' ')
        inorder(node.right)

def preorder(node):
    if node:
        print(node.data, end=' ')
        preorder(node.left)
        preorder(node.right)

def postorder(node):
    if node:
        postorder(node.left)
        postorder(node.right)
        print(node.data, end=' ')

print("Inorder Traversal:")
inorder(root)
print("\nPreorder Traversal:")
preorder(root)
print("\nPostorder Traversal:")
postorder(root)
```

```python
#PENCARIAN NILAI DALAM BST

def search(node, key):
    if node is None or node.data == key:
        return node
    if key < node.data:
        return search(node.left, key)
    return search(node.right, key)

#Uji pencarian
key = 60
result = search(root, key)
if result:
    print(f"{key} ditemukan dalam tree.")
else:
    print(f"{key} tidak ditemukan.")

key = 25
result = search(root, key)
if result:
    print(f"{key} ditemukan dalam tree.")
else:
    print(f"{key} tidak ditemukan.")
```

```python
#Struktur Tree berdasarkan identitas saya

class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

#Root (jumlah huruf nama lengkap) = Dian Khaerunia Azzahra (20 huruf)
root = Node(20)

#Anak kiri root (jumlah vokal) = i, a, a, e, u, i, a, a, a, a (10)
root.left = Node(10)

#Anak kanan root (dua digit terakhir NPM) = 2505060039 (39)
root.right = Node(39)

#Anak kiri dari node kiri (tanggal lahir DD) = 10
root.left.left = Node(10)

#Anak kanan dari node kanan (bulan lahir MM) = 06
root.right.right = Node(6)

#Fungsi inorder traversal
def inorder(node):
    if node:
        inorder(node.left)
        print(node.data, end=' ')
        inorder(node.right)

print("Inorder Traversal Tree Identitas:")
inorder(root)
```

```python
#Binary Search Tree berdasarkan identitas saya

class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

class BinarySearchTree:
    def __init__(self):
        self.root = None

    def insert(self, root, data):
        if root is None:
            return Node(data)
        if data < root.data:
            root.left = self.insert(root.left, data)
        else:
            root.right = self.insert(root.right, data)
        return root

#Data: 3 digit terakhir NPM (039), DD (10), MM (06), ASCII 'D' (68), ASCII 'i' (105), bebas (7, 2)
data_list = [39, 10, 6, 68, 105, 7, 2]

bst = BinarySearchTree()
root = None
for value in data_list:
    root = bst.insert(root, value)

def inorder(node):
    if node:
        inorder(node.left)
        print(node.data, end=' ')
        inorder(node.right)

def preorder(node):
    if node:
        print(node.data, end=' ')
        preorder(node.left)
        preorder(node.right)

def postorder(node):
    if node:
        postorder(node.left)
        postorder(node.right)
        print(node.data, end=' ')

print("=== Traversal BST ===")
print("Inorder   :", end=' ')
inorder(root)
print("\nPreorder  :", end=' ')
preorder(root)
print("\nPostorder :", end=' ')
postorder(root)

def search(node, key):
    if node is None or node.data == key:
        return node
    if key < node.data:
        return search(node.left, key)
    return search(node.right, key)

print("\n\n=== Pencarian ===")

#Cari dua digit terakhir NPM: 39
key = 39
result = search(root, key)
print(f"{key} {'ditemukan' if result else 'tidak ditemukan'} dalam tree.")

#Cari angka yang tidak ada dalam list
key = 99
result = search(root, key)
print(f"{key} {'ditemukan' if result else 'tidak ditemukan'} dalam tree.")
```
